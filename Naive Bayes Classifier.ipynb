```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)
import matplotlib.pyplot as plt

# Generate synthetic dataset
X, y = make_classification(
    n_features=6,
    n_classes=3,
    n_samples=800,
    n_informative=2,
    random_state=1,
    n_clusters_per_class=1,
)

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, marker="*")
plt.title("Generated Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)

# Train Gaussian Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# Single prediction
predicted = model.predict([X_test[6]])
print("Actual Value:", y_test[6])
print("Predicted Value:", predicted[0])

# Evaluate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average="weighted")

print("Accuracy:", accuracy)
print("F1 Score:", f1)

# Confusion Matrix
labels = [0, 1, 2]
cm = confusion_matrix(y_test, y_pred, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot()
plt.title("Confusion Matrix")
plt.show()
```

```python
# Explanation of Code

# 1. Data Generation:
#    - `make_classification` creates a synthetic dataset with 3 classes and 6 features, 2 of which are informative.

# 2. Dataset Visualization:
#    - Plots the first two features to visualize class distribution.

# 3. Train-Test Split:
#    - Splits the dataset into training (67%) and testing (33%) sets.

# 4. Naive Bayes Model:
#    - A Gaussian Naive Bayes** classifier is trained using the `GaussianNB` class.

# 5. Model Evaluation:
#    - Accuracy: Percentage of correct predictions.
#    - F1 Score: Harmonic mean of precision and recall, considering class imbalance.

# 6. Confusion Matrix:
#    - Visualizes the model's performance in predicting each class.

# ---

# Outputs
# - Accuracy: Overall correctness of the model.
# - F1 Score: Weighted average of precision and recall.
# - Confusion Matrix: Distribution of true vs. predicted classes.

# This implementation demonstrates a simple yet effective application of the Naive Bayes classifier for document classification tasks. Let me know if you'd like to extend or customize it!



# Types of Naive Bayes Classifiers

# 1. Gaussian Naive Bayes:
#    - Used for continuous data.
#    - Assumes that the features follow a Gaussian (Normal) distribution.
#    - The probability of a feature (x) given a class (c) is calculated as:
#    - Commonly used in applications like predicting numerical data or continuous-valued features.

# 2. Multinomial Naive Bayes:
#    - Used for discrete data.
#    - Typically applied to text classification (e.g., spam detection, sentiment analysis).
#    - Assumes the features (e.g., word counts) represent frequencies or counts.
#    - The likelihood of the feature given a class is calculated based on the frequency of occurrences.

# 3. Bernoulli Naive Bayes:
#    - Used for binary data.
#    - Assumes that features are binary-valued (e.g., presence or absence of a word in a document).
#    - Useful in text classification where the focus is on the presence/absence of terms rather than their frequency.

# ---

# Confusion Matrix

# A confusion matrix is a tool to evaluate the performance of a classification model by comparing true labels with predicted labels. It summarizes the prediction results into a table with the following components:

# |                  | Predicted Positive | Predicted Negative |
# |------------------|-------------------------|-------------------------|
# | True Positive | Correctly predicted positives |
# | True Negative | Correctly predicted negatives |
# | False Positive | Incorrectly predicted positives |
# | False Negative | Incorrectly predicted negatives |


# Metrics Derived from the Confusion Matrix:

# 1. Accuracy: 
#    - Proportion of correct predictions.
   
#      Accuracy = TP + TN/TP + TN + FP + FN

# 2. Precision (Positive Predictive Value): 
#    - Proportion of true positive predictions out of all positive predictions.
     
#      Precision = TP/TP + FP
   
# 3. Recall (Sensitivity or True Positive Rate): 
#    - Proportion of actual positives that are correctly identified.
     
#      Recall = TP/TP + FN

# 4. F1-Score:
#    - Harmonic mean of precision and recall.
   
#      F1 = 2 * Precision * Recall / Precision + Recall
   
# ---


# 3 class classification problem:

# |                | Predicted: 0 | Predicted: 1 | Predicted: 2 |
# |----------------|------------------|------------------|------------------|
# | True: 0    | 50               | 2                | 1                |
# | True: 1    | 3                | 45               | 7                |
# | True: 2    | 4                | 6                | 43               |


# From this:
# - Accuracy: The total correct predictions divided by all predictions.
# - Per-Class Precision/Recall: Computed for each class individually, useful for imbalanced datasets.

# Accuracy: Overall correctness of the model.
# F1 Score: Weighted average of precision and recall.
# Confusion Matrix: Distribution of true vs. predicted classes.
```
