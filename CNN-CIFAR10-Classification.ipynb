```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score
import seaborn as sns
import numpy as np

# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

# One-hot encode the labels
train_labels = to_categorical(train_labels, num_classes=10)
test_labels = to_categorical(test_labels, num_classes=10)

# Build the CNN model
model = models.Sequential()

# Convolutional layers
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten layer
model.add(layers.Flatten())

# Dense layers
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Display the model summary
model.summary()

# Train the model
history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))

# Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'\nTest accuracy: {test_acc:.4f}')

# Get predictions
y_pred = np.argmax(model.predict(test_images), axis=1)
y_true = np.argmax(test_labels, axis=1)

# Compute confusion matrix, precision, recall
conf_matrix = confusion_matrix(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='macro')
recall = recall_score(y_true, y_pred, average='macro')
accuracy = accuracy_score(y_true, y_pred)

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'Accuracy: {accuracy:.4f}')

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Plot training history
plt.figure(figsize=(12, 4))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')

plt.tight_layout()
plt.show()
```

```python
1. Convolutional Neural Network (CNN) Overview:
A Convolutional Neural Network (CNN) is a type of deep learning model specifically designed for image data. CNNs can automatically and adaptively learn spatial hierarchies of features through backpropagation. 
CNNs are widely used in tasks like image classification, object detection, and face recognition.


1. Dataset Loading and Preprocessing:

Dataset: CIFAR-10 is a popular image classification dataset that contains 60,000 color images (32x32 pixels) in 10 different classes (airplanes, cars, birds, etc.).
Normalization: Pixel values in images range from 0 to 255. By dividing them by 255, we scale the values to be between 0 and 1. This helps with faster and more stable convergence during training.
One-hot Encoding: The labels are converted to one-hot encoded vectors, meaning each label is represented as a 10-element vector with a 1 in the position corresponding to the class index and 0s elsewhere.

2. Building the CNN Model:

1. Input Shape: The shape of each image is (32, 32, 3) where 32x32 represents the width and height, and 3 represents the color channels (RGB).

b. Convolutional Layer:

Convolutional Layer: A convolutional layer applies filters (kernels) to the input image to detect features like edges, textures, etc. 
Each filter slides over the input, performing an element-wise multiplication and summing the result (this operation is called a convolution).

Filters/Channels: The 32 here represents the number of filters (output feature maps) that the layer learns. Each filter is of size 3x3.

Activation Function (ReLU): The activation='relu' specifies the use of the Rectified Linear Unit (ReLU) activation function. ReLU is defined as f(x) = max(0, x). 
It introduces non-linearity, making the model capable of learning complex patterns.

c. Pooling Layer: 

Max Pooling Layer: This layer reduces the spatial dimensions of the input (i.e., width and height) by taking the maximum value in each 2x2 window of the input. 

model.add(layers.MaxPooling2D((2, 2)))

This helps in: Reducing the computational load.

d. Second Convolutional and Pooling Layer:

This layer performs another round of convolution and pooling to further learn more complex features from the input. The number of filters is increased to 64, which allows the model to capture more patterns.Preventing overfitting.
Making the model more robust to small translations of the input.

This layer performs another round of convolution and pooling to further learn more complex features from the input. The number of filters is increased to 64, which allows the model to capture more patterns.

e. Flatten Layer:

model.add(layers.Flatten())

Flatten Layer: After the convolution and pooling layers, the output is a multi-dimensional tensor (e.g., [batch_size, height, width, channels]). 
The Flatten layer reshapes this multi-dimensional tensor into a 1D vector so it can be passed into fully connected (dense) layers.

f. Dense Layer:

model.add(layers.Dense(128, activation='relu'))

Dense Layer (Fully Connected Layer): This layer is a standard neural network layer where each neuron is connected to every neuron in the previous layer. In this case, the dense layer has 128 neurons.
Activation (ReLU): Again, ReLU is used here to introduce non-linearity.

g. Dropout Layer:

model.add(layers.Dropout(0.5))

Dropout: This layer randomly "drops out" (sets to zero) 50% of the neurons in the layer during training. 
This is a regularization technique that helps prevent overfitting by ensuring that the model doesn't become too reliant on any single neuron.

h. Output Layer:

model.add(layers.Dense(10, activation='softmax'))

Dense Layer (Output): This is the final layer of the network, with 10 neurons (one for each class in the CIFAR-10 dataset).

Softmax Activation: The softmax function outputs a probability distribution over the 10 classes, ensuring that the sum of all output values equals 1. 
The class with the highest probability is taken as the predicted class.

3. Compiling the Model:

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

Optimizer (Adam): Adam (Adaptive Moment Estimation) is a popular optimization algorithm that adapts the learning rate for each parameter. 

It’s widely used because it combines the benefits of two other optimizers: AdaGrad and RMSProp.

Loss Function (Categorical Crossentropy): This loss function is used for multi-class classification problems.

It measures how far the predicted probability distribution is from the true distribution (one-hot encoded labels).

Metrics (Accuracy): The model will be evaluated based on accuracy, which is the ratio of correctly classified images to the total number of images.

4. Training the Model:

history = model.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))

Epochs: The model will train for 10 epochs, meaning it will pass through the entire dataset 10 times.

Validation Data: The test_images and test_labels are used as validation data, meaning the model’s performance is evaluated on this data after each epoch, but it is not used for training.

                            
5. Evaluating the Model:

test_loss, test_acc = model.evaluate(test_images, test_labels)

print(Test accuracy: {test_acc:.4f}

After training, the model is evaluated on the test data to see how well it performs on unseen data. The test accuracy is printed as the final result.

Summary of Layers:
                            
Convolutional Layer: Detects features (edges, textures, etc.) by convolving filters over the input image.
Activation (ReLU): Introduces non-linearity to enable learning of complex patterns.
Pooling Layer: Reduces spatial dimensions while retaining important features.
Flatten Layer: Converts multi-dimensional data into a 1D vector for the dense layers.
Dense Layer: Fully connected layers that make predictions based on learned features.
Dropout: Regularizes the model by randomly dropping neurons to prevent overfitting.
Softmax: Outputs probabilities for each class.

Explanation of Metrics:

Confusion Matrix: Shows the performance of the classification by comparing the predicted classes with the true classes.
Precision: The ratio of correctly predicted positive observations to the total predicted positives.
Recall: The ratio of correctly predicted positive observations to the actual positives.
Accuracy: The ratio of correctly predicted instances to the total instances.
```
